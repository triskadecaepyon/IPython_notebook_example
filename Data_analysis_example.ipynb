```
mu1 <- 0   
mu2 <- 0	
sig1 <- 2	# From Var(x)=4
sig2 <- 3	# From Var(y)=5
rho <- 0	# corr(x, y)

xm <- -3
xp <- 3
ym <- -3
yp <- 3

x <-seq(-8,8, length.out = 100)
y <-seq(-8,8, length.out = 100)

# Bivariate function
# Adapted from https://quantcorner.wordpress.com/2012/09/21/bivariate-normal-distribution-with-r/
bivariate <- function(x,y){
  term1 <- 1 / (2 * pi * sig1 * sig2 * sqrt(1 - rho^2))
  term2 <- (x - mu1)^2 / sig1^2
  term3 <- -(2 * rho * (x - mu1)*(y - mu2))/(sig1 * sig2)
  term4 <- (y - mu2)^2 / sig2^2
  z <- term2 + term3 + term4
  term5 <- term1 * exp((-z / (2 *(1 - rho^2))))
  return (term5)
}

# Computes the density values
z <- outer(x,y,bivariate)
contour(x, y, z, main="Contours for Bivariate Data")
```

c)  
The Bivariate is described by the following equation: $f(x,y) = \frac{1}{2 \pi \sigma _{x} \sigma _{y} \sqrt{1-{\rho}^2}} exp $
$ \frac{-1}{2(1-\rho^2)} [\frac{(x-\mu _{x})^2}{\sigma _{x}^{2}} + \frac{(y-\mu _{y})^2}{\sigma _{y}^{2}} - \frac{2 \rho (x-\mu _{x})(y-\mu _{y})}{\sigma _{x}^{2} \sigma _{y}^{2}}]$  
  
When x is set to 1, the formuala becomes:
  
$f(1,y) = \frac{1}{2 \pi (2) (3) \sqrt{1-{.5}^2}} exp $
$ \frac{-1}{2(1-(.5)^2)} [\frac{(1-0)^2}{4} + \frac{(y-0)^2}{9} - \frac{2 (.5) (1-0)(y-0)}{(4)(9)}]$  
  
Reducing to:  
  
$f(1,y) = \frac{1}{6 \pi} exp $
$ \frac{-1}{1.5} [\frac{4 (y)^2 - y + 9}{36}]$  
  
When y is set to 1, the formuala becomes:
  
$f(x,1) = \frac{1}{2 \pi (2) (3) \sqrt{1-{.5}^2}} exp $
$ \frac{-1}{2(1-(.5)^2)} [\frac{(x-0)^2}{4} + \frac{(1-0)^2}{9} - \frac{2 (.5) (x-0)(1-0)}{(4)(9)}]$  
  
Reducing to:  
  
$f(x,1) = \frac{1}{6 \pi} exp $
$ \frac{-1}{1.5} [\frac{9x^2 - x + 4}{36}]$  
  
So the final two answers are: 
  
$f(1,y) = \frac{1}{6 \pi} exp $
$ \frac{-1}{1.5} [\frac{4 (y)^2 - y + 9}{36}]$  
  
$f(x,1) = \frac{1}{6 \pi} exp $
$ \frac{-1}{1.5} [\frac{9x^2 - x + 4}{36}]$

**MLR using Scikit Learn**

*Part a) LSTAT and MEDV Box plots*

````python
from matplotlib import pyplot as plt
%matplotlib inline
from sklearn import datasets
import seaborn as sns
import numpy as np

datasets.load_boston()
bost_data = datasets.load_boston()
b = bost_data.data.T
b_lstat=np.copy(b[12])
plt.title("LSTAT")
plt.boxplot(b_lstat)
````

Outliers for the LSTAT are defined by Q1 - 1.5IQR, and Q3 + 1.5IQR.
The calculation for this is below:

````python
#Let's generate the IQR's and ranges first
q75 = np.percentile(b_lstat, 75)
q25 = np.percentile(b_lstat, 25)
iqr = q75 - q25
print "Q3 " + str(q75)
print "Q1 " + str(q25)
print "IQR " + str(iqr)
print "Outlier 1: " + str(q75 + 1.5*iqr)
print "Outlier 2: " + str(q25 - 1.5*iqr)
````

*Now for the MEDV (Target) Data:*

````python
plt.title("MEDV")
plt.boxplot(bost_data.target)
````

````python
q75_MEDV = np.percentile(bost_data.target, 75)
q25_MEDV = np.percentile(bost_data.target, 25)
iqr_MEDV = q75_MEDV - q25_MEDV
print "Q3 " + str(q75_MEDV)
print "Q1 " + str(q25_MEDV)
print "IQR " + str(iqr_MEDV)
print "Outlier 1: " + str(q75_MEDV + 1.5*iqr_MEDV)
print "Outlier 2: " + str(q25_MEDV - 1.5*iqr_MEDV)
````

*Part b) LSTAT and MEDV Scatter Plots and outliers*

````python
plt.title("MEDV vs. LSTAT, Raw data")
plt.scatter(bost_data.target, b_lstat)
````

Now we will create the new, clean set of data that we will overlay.  This requires taking the outliers out of the original dataset, which we will replace with NaN in the numpy array to preserve statistical ordering.

````python
new_medv = np.copy(bost_data.target)
for i in range(0,new_medv.size):
    if new_medv[i] > (q75_MEDV + 1.5*iqr_MEDV) or new_medv[i] < (q25_MEDV - 1.5*iqr_MEDV):
        new_medv[i] = np.NAN
        
new_lstat = np.copy(b_lstat)
for i in range(0,b_lstat.size):
    if new_lstat[i] > (q75 + 1.5*iqr) or new_lstat[i] < (q25 - 1.5*iqr):
        new_lstat[i] = np.NAN
````

````python
plt.title("MEDV vs. LSTAT, Clean vs. Raw data")
plt.scatter(bost_data.target, b_lstat,color="red", label="Dataset Outliers")
plt.scatter(new_medv, new_lstat, color="blue", label="Within 1.5 IQR")
plt.legend(loc='upper right')
````

*Part C) MLR with ScikitLearn*

````python
# Split the datasets into different parts, training and test
#CRIM bostrain[0,]
#ZN bostrain[1,]
#CHAS bostrain[3]
#RM bostrain[5,]
#LSTAT bostrain[12,]

bostrain = b[[0,1,3,5,12],:350]
bostest = b[[0,1,3,5,12],350:]

#Take the log of the medv data, and split
lmedv = np.log(np.copy(bost_data.target))
lmedv_train = lmedv[:350]
lmedv_test = lmedv[350:]

````

````python
from sklearn import linear_model
regr = linear_model.LinearRegression()

regr.fit(bostrain.T, lmedv_train)
lr_y = regr.predict(bostest.T)

from sklearn.metrics import mean_squared_error
mse = mean_squared_error(lmedv_test, lr_y)
print "MSE on bostest: " +str(mse)

lr_y_2 = regr.predict(bostrain.T)
mse_2 = mean_squared_error(lmedv_train, lr_y_2)

print "MSE on bostrain: " + str(mse_2)
````

i) The mean squared error (MSE) obtained using bostrain's data is **0.0176**, which increases by **0.1483** to 0.1658 on bostest
